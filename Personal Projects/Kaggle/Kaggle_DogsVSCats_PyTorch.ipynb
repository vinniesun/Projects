```python
import os, shutil

name = './DogsVsCats/train/'
training_image_length = len(os.listdir(name))
training_image_length
```

```python
# Root Directory
root_dir = './DogsVsCats'

# Directory for training & Testing
train_dir = os.path.join(root_dir, 'train')
test_dir = os.path.join(root_dir, 'test1')

# Directory for training cat images
cat_train_dir = os.path.join(train_dir, 'cats')
#os.mkdir(cat_train_dir)
# Directory for training dog images
dog_train_dir = os.path.join(train_dir, 'dogs')
#os.mkdir(dog_train_dir)

# store all instances of cat images in cat list and dog images in dog list
cat_names = ['cat.{}.jpg'.format(i) for i in range(training_image_length+1) if os.path.isfile('./DogsVsCats/train/' + 'cat.{}.jpg'.format(i))]
dog_names = ['dog.{}.jpg'.format(i) for i in range(training_image_length+1) if os.path.isfile('./DogsVsCats/train/' + 'dog.{}.jpg'.format(i))]
len(cat_names), len(dog_names)
```

```python
# copy all files in cat_names into the cats training folder
for name in cat_names:
  src = os.path.join(train_dir, name)
  dst = os.path.join(cat_train_dir, name)
  shutil.move(src, dst)

for name in dog_names:
  src = os.path.join(train_dir, name)
  dst = os.path.join(dog_train_dir, name)
  shutil.move(src, dst)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import torch
import torch.nn as nn
import cv2
import matplotlib.pyplot as plt
import torchvision
from torch.utils.data import Dataset, DataLoader, ConcatDataset
from torchvision import transforms
%matplotlib inline
```

```python
data_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.ColorJitter(),
    transforms.RandomCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.Resize(128),
    transforms.ToTensor()
])
```

```python
# Cats & Dogs training data setup
# According to the rules of the DataLoader in pytorch you should choose the the superior path of the image path. 
# That means if your images locate in './Dataset/images/', the path of the data loader should be './Dataset' instead
train_data = torchvision.datasets.ImageFolder(
    root=train_dir,
    transform=data_transform
)
```

```python
test_imgs = ['{}.jpg'.format(i) for i in range(1, len(os.listdir(test_dir)))]
print(len(test_imgs))
#os.mkdir(test_dir+'/test')

for name in test_imgs:
  src = os.path.join(test_dir, name)
  dst = os.path.join(test_dir+'/test', name)
  shutil.move(src, dst)
```

```python
# Cats & Dogs test data setup
test_data = torchvision.datasets.ImageFolder(
    root=test_dir,
    transform=data_transform
)
```

```python
# Setup train_loader and test_loader
train_loader = torch.utils.data.DataLoader(
    dataset=train_data,
    batch_size=32,
    shuffle=True,
    drop_last=True
)

test_loader = torch.utils.data.DataLoader(
    dataset=test_data,
    batch_size=32,
    shuffle=False
)
```

```python
samples, labels = iter(train_loader).next()
plt.figure(figsize=(16,24))
grid_imgs = torchvision.utils.make_grid(samples[:24])
np_grid_imgs = grid_imgs.numpy()
# In a PyTorch tensor, image is (batch, width, height).
# To use this in Numpy fashion, you have to transpose it to (width, height, batch) to show it.
plt.imshow(np.transpose(np_grid_imgs, (1, 2, 0)))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model = torchvision.models.resnet18(pretrained=True)
```

```python
print(model)
```

```python
# Freeze the parameters so we don't backprop through them during training
for param in model.parameters():
  param.requires_grad = False
```

```python
# Define our new Classifier
input_features = model.fc.in_features
#input_features = 25088

new_classifier = nn.Sequential(
    nn.Linear(input_features, 256),
    nn.ReLU(),
    nn.Dropout2d(p=0.5),
    nn.Linear(256, 2)
)

model.fc = new_classifier
```

```python
criterion = nn.CrossEntropyLoss()
optimiser = torch.optim.Adam(model.parameters(), lr=0.002)
#scheduler = torch.optim.lr_scheduler.StepLR(optimiser, step_size=5, gamma=0.1)
```

```python
epochs = 20
total_step = len(train_loader)
model.to(device)
loss_list = []
acc_list = []
total_loss = 0
per_iteration = 100

for epoch in range(epochs):
  for i, (images, labels) in enumerate(train_loader):
    images = images.to(device)
    labels = labels.to(device)

    # Forward Pass
    outputs = model(images)
    loss = criterion(outputs, labels)

    # Backward and optimise
    optimiser.zero_grad()
    loss.backward()
    optimiser.step()
    total_loss += loss.item()
    #scheduler.step()

    # Validation
    #pred = torch.argmax(outputs, dim=1)
    #correct = pred.eq(labels)
    #total_acc += torch.mean(correct.float())

    if (i+1)%100 == 0:
      pred = torch.argmax(outputs, dim=1)
      correct = pred.eq(labels)
      acc = torch.mean(correct.float())
      print('Epoch: {}/{}, Step: {}/{}, Loss: {:.4f}, Accuracy: {:.4}'.format(epoch+1, epochs, i+1, total_step, total_loss/per_iteration, acc))
      loss_list.append(total_loss/per_iteration)
      acc_list.append(acc)
      total_loss = 0

plt.plot(loss_list, label='loss')
plt.plot(acc_list, label='accuracy')
plt.legend()
plt.title('Training loss and Accuracy')
plt.show 
```

The random spikes in the loss plot is caused by the mini-batch gradient descent in Adam(Batch_size=32). This is because some mini-batches have **'by chance'** unlucky data for the optimisation.<br>
Loglikelihood-losses like CrossEntropyLoss(logSoftmax+NLLLoss) needs to be clipped, as it may evaluate near log(0) for bad predictions/outliers in dataset, causing exploding gradients.
